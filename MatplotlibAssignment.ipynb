```python
#Load the dataset into a Pandas DataFrame and explore the data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = ("C:/Users/user/Documents/Africa Data School/Matplotlib/Jumta.csv")

df = pd.read_csv(data)

#Create a line plot showing the monthly revenue over the two-year period. Customize the plot to make it publication-quality.
fig, ax = plt.subplots()

ax.plot(df["Month"],df["Revenue"], marker='o', label ="Revenue")
ax.set_title("Month Revenue 2020-2021")
ax.set_xlabel("Month") 
ax.set_ylabel("Revenue")
plt.xticks(rotation=45)
ax.legend()
plt.show()
```

```python
#Create a bar plot showing the total number of orders per month. Customize the plot to make it publication-quality

plt.bar(df["Month"], df["Orders"], label="Orders")
plt.title("Total number of orders per month")
plt.xlabel("Month")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
#Create a scatter plot showing the relationship between the number of orders and the average order value. Customize the plot to make it publication-quality
plt.scatter(df["Orders"], df["Average_Order_Value"], label="Average Order Value")
plt.title("Number of Orders vs Average Order Value")
plt.xlabel("Orders")
plt.ylabel("Order Value")
plt.legend()
plt.show()
```

```python
#Create a subplot containing all three visualizations created in steps 2-4. Adjust the layout and size of the subplot to make it visually appealing
fig, (ax1, ax2, ax3) = plt.subplots(nrows=3, ncols=1,figsize=(10, 12))
#line plot
ax1.plot(df["Month"],df["Revenue"], color="green", marker='o', label ="Revenue")
ax1.set_title("Month Revenue 2020-2021")
ax1.set_xlabel("Month") 
ax1.set_ylabel("Revenue")
ax1.tick_params(axis='x', rotation=45)
ax1.legend()
#bar plot
ax2.bar(df["Month"],df["Orders"], color="green", label ="Orders")
ax2.set_title("Total number of orders per month")
ax2.set_xlabel("Month") 
ax2.set_ylabel("Number of Orders")
ax2.tick_params(axis='x', rotation=45)
ax2.legend()
#scatter plot
ax3.scatter(df["Orders"],df["Average_Order_Value"], color="green", label ="Average Order Value")
ax3.set_title("Number of Orders vs Average Order Value")
ax3.set_xlabel("Orders") 
ax3.set_ylabel("Order Value")
ax3.legend()
plt.tight_layout()
plt.savefig('jumpta_sales_analysis.png', dpi=500)
plt.show()
```
